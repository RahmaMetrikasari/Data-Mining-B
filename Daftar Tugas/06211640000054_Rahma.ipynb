```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
# memanggil data

data=pd.read_csv('online_retail1.csv')
```

## Preprocesing Data

```python
# menghapus missing value pada variabel describe

data['Description'] = data['Description'].str.strip()
data.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
```

```python
# menghapus pembelian kredit pada variabel Invoiceno dengan awalan c

data['InvoiceNo'] = data['InvoiceNo'].astype('str')
data = data[~data['InvoiceNo'].str.contains('C')]
```

```python
# menghapus missing value pada describe

data.dropna(axis=0, subset=['Description'], inplace=True)
```

```python
# menghapus missing value pada variabel customer id

data.dropna(axis=0, subset=['CustomerID'], inplace=True)
```

```python
# menghapus missing value pada data country

data.dropna(axis=0, subset=['Country'], inplace=True)
```

```python
# mengimport hasil ke excel untuk dilihat masih ada missing value atau tidak 

data.to_excel('data_new.xlsx',index=0)
```

## Association Rule

```python
# mengetahui jumlah negara pada variabel country

data['Country'].value_counts()
```

```python
# mengetahui jumlah item pada variabel description

data['Description'].value_counts()
```

```python
# mengelompokan transaksi berdasarkan nomor nota
y = data.groupby('InvoiceNo')
transaksi = []
for i,j in y:
    transaksi.append(list(j['Description'].map(str)))
datatrans = pd.DataFrame(transaksi)
datatrans.head()
```

```python
# membuat array untuk mennyimpan association rule

records=[]
for i in range (0, 18528):
    records.append([str(datatrans.values[i,j]) for j in range (0,542)])
```

```python
association_rules=apriori(records, min_support=0.005,
                         min_confidence=0.2,
                         min_lift=3, min_length=2)
association_result=list(association_rules)
```

```python
association_result[0]
```

```python
for item in association_result:
#Index Pertama Untuk Inner List
#Base Items & Add Items
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items [0] + " -> " + items [1])
    
    #Index Kedua Untuk Inner List
    print("Support: " + str(item[1]))
    
    #Index Ketiga Untuk List ke-10
    #Index Ketiga Inner List
    
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
