```python
#mengimport pandas dan numpy

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#memanggil data

data = pd.read_excel('datmin.xlsx')
data1 = pd.read_excel('dtm.xlsx')

#memunculkan data 
print(data)
print(data1)

#menggabungkan dua data frame menjadi satu berdasarkan variabel tertentu menjadi data frame baru
data_gab = pd.merge(data,data1,on ='Nama Dosen')
```

```python
#menambahkan kolom keterangan aktif jika score sinta 3 tahun >0
data_gab['Keterangan']=np.where(data_gab['sinta_score_3_year']>0,"Aktif","Tidak")
print(data_gab)
```

```python
#menghitung proporsi yang aktif dengan mengcoding 
data_gab['Code'] = np.where(data_gab['Keterangan']=='Aktif',1,0)
print(data_gab)
proporsi = data_gab.loc[:,'Code'].mean()*100
print('Proporsi Dosen Aktif = ',proporsi,'%')
```

```python
#mengeksploretabel

#menghitung statistika deskriptif 

rata1 = data_gab.loc[:,'scopus_hindex'].mean()
rata2 = data_gab.loc[:,'sinta_score'].mean()
rata3 = data_gab.loc[:,'scopus_citation'].mean()
rata4 = data_gab.loc[:,'sinta_score_3_year'].mean()
rata5 = data_gab.loc[:,'google_citations'].mean()
rata6 = data_gab.loc[:,'google_hindex'].mean()

std1 = data_gab.loc[:,'scopus_hindex'].std()
std2 = data_gab.loc[:,'sinta_score'].std()
std3 = data_gab.loc[:,'scopus_citation'].std()
std4 = data_gab.loc[:,'sinta_score_3_year'].std()
std5 = data_gab.loc[:,'google_citations'].std()
std6 = data_gab.loc[:,'google_hindex'].std()

max1 = data_gab.loc[:,'sinta_score_3_year'].max()
min1 = data_gab.loc[:,'sinta_score_3_year'].min()

print('rata-rata Skor SINTA 3 Tahun  = ',rata4)
print('rata-rata Skor SINTA = ',rata2)
print('rata-rata Google Citation = ',rata5)
print('rata-rata Scopus H index = ',rata1)
print('rata-rata Google H index = ',rata6)
print('rata-rata SINTA score= ',rata2)

print('Nilai SINTA 3 Tahun Tertinggi  = ',max1)
print('Nilai SINTA 3 Tahun Terendah  = ',min1)

print('Simpangan Baku Skor SINTA 3 Tahun  = ',std4)
print('Simpangan Baku Skor SINTA = ',std2)
print('Simpangan Baku Google Citation = ',std5)
print('Simpangan Baku Scopus H index = ',std1)
print('Simpangan Baku Google H index = ',std6)
print('Simpangan Baku SINTA score= ',std2)



#membuat plot berdasarkan nilai pada tabel
y = data_gab['google_citations']
x = data_gab['google_hindex']

plt.scatter(x,y,c=y,cmap='viridis')
plt.colorbar()
plt.show


```
