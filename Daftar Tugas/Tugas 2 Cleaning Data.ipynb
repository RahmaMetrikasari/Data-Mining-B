# Melakukan Pre-Processing Data

Mengatasi missing Value

```python
# mengetahui jumlah missing value tiap variabel 
jumlah_missing_value = pd.Series(data.isnull().sum())
proporsi_missing_value = pd.Series(100*(data.isnull().sum())/len(data))
data_mis = pd.concat([jumlah_missing_value,proporsi_missing_value],axis=1)
data_mis_val = data_mis.rename(columns = {0 : 'Jumlah Missing Values', 1 : 'Proporsi Missing Value', 2 :'Keterangan'})

data_mis_val
```

```python
# menghapus variabel yang proporsi missing value nya > 30
data_=data.drop('STDs: Time since first diagnosis',axis=1)
data_new=data_.drop('STDs: Time since last diagnosis',axis=1)
data_new
```

```python
# mengatasi missing value
#imputasi data kategorik dengan modus sesuai kolom masing2
data_new = data_new.apply(lambda x:x.fillna(x.value_counts().index[0]))
data_new.isnull().any()
data_new.head()
```

```python
#imputasi data numerik dengan mean sesuai kolom masing-masing
data_new['Number of sexual partners'] = data_new['Number of sexual partners'].fillna((data_new['Number of sexual partners'].mean()))
data_new['First sexual intercourse'] = data_new['First sexual intercourse'].fillna((data_new['First sexual intercourse'].mean()))
data_new['Num of pregnancies'] = data_new['Num of pregnancies'].fillna((data_new['Num of pregnancies'].mean()))
data_new['Smokes (years)'] = data_new['Smokes (years)'].fillna((data_new['Smokes (years)'].mean()))
data_new['Hormonal Contraceptives (years)'] = data_new['Hormonal Contraceptives (years)'].fillna((data_new['Hormonal Contraceptives (years)'].mean()))
data_new['IUD (years)'] = data_new['IUD (years)'].fillna((data_new['IUD (years)'].mean()))
data_new['STDs (number)'] = data_new['STDs (number)'].fillna((data_new['STDs (number)'].mean()))

data_new
```

# Melakukan Uji Multivariat Normal dengan Shapiro Wilk Test

```python
# data yang digunakan hanya data numerik
dataa=data_new[['Age','Number of sexual partners','First sexual intercourse','Num of pregnancies','Smokes (years)','Hormonal Contraceptives (years)','IUD (years)','STDs (number)']]
dataa
```

```python
from numpy.random import seed
from numpy.random import randn
from scipy.stats import shapiro
# normality test
stat, p = shapiro(dataa)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Gagal Tolak H0, data berdistribusi normal multivariat')
else:
	print('Tolak H0, data tidak berdistribusi normal multivariat')
```
